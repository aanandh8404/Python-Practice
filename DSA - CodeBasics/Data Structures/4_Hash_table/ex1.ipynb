```python
# nyc_weather.csv contains new york city weather for first few days in the month of January. Write a program that can answer following,
# Figure out data structure that is best for this problem

arr = []
with open('nyc_weather.csv','r') as fp:
    for line in fp:
        element = line.split(",")

        try:
            temperature = int(element[1])
            arr.append(temperature)
        except:
            pass

print(arr)
```

```python
# What was the average temperature in first week of Jan
print(sum(arr[0:7])/7)
```

```python
# What was the maximum temperature in first 10 days of Jan
print(max(arr[0:10]))
```

```python
weather_dict = {}
with open('nyc_weather.csv','r') as fp:
    for line in fp:
        elem = line.split(",")
        try:
            temp = int(elem[1])
            weather_dict[elem[0]] = temp
        except:
            pass

print(weather_dict)
```

```python
word_count = {}

with open('poem.txt','r') as fp:
    for line in fp:
        words = line.split(" ")
        for word in words:
            word = word.replace('\n','')
            if word in word_count:
                word_count[word] += 1
            else:
                word_count[word] = 1

print(word_count)
```
